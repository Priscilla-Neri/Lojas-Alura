### Importa√ß√£o dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. An√°lise do faturamento


```python
faturamento_loja1 = loja['Pre√ßo'].sum()
faturamento_loja2 = loja2['Pre√ßo'].sum()
faturamento_loja3 = loja3['Pre√ßo'].sum()
faturamento_loja4 = loja4['Pre√ßo'].sum()


# Lista com os faturamentos
lista_faturamento = [
    ('Loja 1', faturamento_loja1),
    ('Loja 2', faturamento_loja2),
    ('Loja 3', faturamento_loja3),
    ('Loja 4', faturamento_loja4)
]

# Encontrar a loja com menor faturamento
menor_faturamento = min(lista_faturamento, key=lambda x: x[1])

# Resultado
print(f' üü• A loja com menor faturamento √©: {menor_faturamento[0]} com R${menor_faturamento[1]:.2f}')
```

```python
faturamento_loja1 = loja['Pre√ßo'].sum()
print(f'R${faturamento_loja1:.2f}')

faturamento_loja2 = loja2['Pre√ßo'].sum()
print(f'R${faturamento_loja2:.2f}')

faturamento_loja3 = loja3['Pre√ßo'].sum()
print(f'R${faturamento_loja3:.2f}')

faturamento_loja4 = loja4['Pre√ßo'].sum()
print(f'R${faturamento_loja4:.2f}')
```

```python
# Carregar os dados
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Adicionando coluna identificadora de loja
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Juntando todas as lojas
todas_lojas = pd.concat([loja1, loja2, loja3, loja4])
```

```python
# Tabela de quantidade de produtos vendidos por loja e categoria
tabela_final = todas_lojas.groupby(['Loja', 'Categoria do Produto'])['Produto'].count().unstack(fill_value=0)

# Transpor para facilitar o gr√°fico horizontal
tabela_plot = tabela_final.T
```

```python
# Plotar o gr√°fico
tabela_plot.plot(kind='barh', figsize=(12, 6), colormap='tab10')

# T√≠tulos e r√≥tulos
plt.title('Quantidade de Produtos Vendidos por Categoria e Loja', fontsize=14, fontweight='bold')
plt.xlabel('Quantidade de Produtos')
plt.ylabel('Categorias')

# Legenda posicionada abaixo
plt.legend(title='Lojas', bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=4)

# Linhas de grade discretas
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
